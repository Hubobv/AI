```python
import pandas as pd
url='/passenger.csv'
data=pd.read_csv(url)
data.head()

import matplotlib.pyplot as plt
data.hist()
plt.show()

correlation=data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot=True)
sns.pairplot(data)
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
#learning data
model=linear_model.LinearRegression()
model.fit(X=data.loc[:,['Passenger','Miles(X)']],y=data['Dollars (Y)'])# 2 bien
print('intercept la: ',model.intercept_)
print('coeffient la: ',model.coef_)
print('do chinh xac: ',model.score(X=data.loc[:,['Passenger','Miles(X)']],y=data['Dollars (Y)']))

#bac 5
model=linear_model.LinearRegression()
X=pd.DataFrame([data['Miles(X)'],data['Miles(X)']**2,data['Miles(X)']**3,data['Miles(X)']**4,data['Miles(X)']**5]).T 
model.fit(X,y=data['Dollars (Y)'])
print("Intercept bac 5 cua mo hinh:", model.intercept_)
print("Coeffient bac 5 cua mo hinh:", model.coef_)
print('Do chinh xac bac 5 cua mo hinh:',model.score(X,y=data['Dollars (Y)'])) 
```

```python
import numpy as np
range= np.arange(0,5500,1)
X_range=pd.DataFrame([range,range**2,range**3,range**4,range**5]).T 
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='Miles(X)',
          y='Dollars (Y)',figsize=(9,9),color='black',
          xlim=(0,5500))
plt.plot(range,y_pred,color='blue')
```
